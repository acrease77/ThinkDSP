# Question 6

```python
class SawtoothSignal_Alex(thinkdsp.Sinusoid):
    """Represents a Sawtooth signal."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        cycles = self.freq * ts + self.offset / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = frac
        ys = thinkdsp.normalize(thinkdsp.unbias(ys), self.amp)
        return ys
```

```python
sig=SawtoothSignal_Alex()
wave=sig.make_wave(duration=2, start=0, framerate=20000)
seg=wave.segment(start, duration=0.003)
seg.plot()
```

```python
seg2=wave.segment(start, duration=0.1)
seg2.plot()
```

The sawtooth wave is basically generated, but as you can see by the second plot, there is a small offset every few cycles. I'm not sure why it's doing this!

# Question 9

```python
square = thinkdsp.SquareSignal(freq = 1100, amp = 1)
squarewave=square.make_wave(duration=1, framerate=10000)
spectrum = squarewave.make_spectrum()
spectrum.plot()
```

```python
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 50:
        print(f, amp)
```

There are a lot more smaller harmonics in the square wave frequencies than in the triangle wave. The fundamental frequencies are not the dominant frequencies in either case. In the triangle wave signal, all the harmonics are multiples of the dominant frequency. In the square wave, the fundamental frequency is about 100 Hz, and there is a harmonic at every addition of 200 Hz.
