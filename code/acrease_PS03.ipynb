## Question 1

Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
nrows = 1600
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
wave.plot()
```

```python
spec = wave.make_spectrum()
spec.plot()
thinkplot.config(xlabel = 'frequency', ylabel='Closing Price', ylim = [0,100000], xlim = [0,.1],legend=False)
```

Now plotted on a log-log scale:

```python
spec.plot()
thinkplot.config(xlabel = 'frequency', ylabel='Closing Price', ylim = [0,100000], xlim = [0,.1],legend=False, xscale='log', yscale='log')
```

The spectrum looks like red noise because the power is generally decreasing as the frequency increases, and it doesn't deacrease linearly like Brownian noise.

## Question 2

Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
lags, corrs = autocorr.autocorr(wave)
thinkplot.plot(lags,corrs)
thinkplot.config(xlabel = 'Correlation', ylabel='Lag',legend=False)
```

There is no evidence of periodic behavior because then we would see reocccuring peaks. The autocorrelation function is decaying, but not exponentially.

Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch?

```python
lags, corrs = autocorr.autocorr(saxseg)
thinkplot.plot(lags,corrs)
thinkplot.config(xlabel = 'Correlation', ylabel='Lag',legend=False)
```

```python
low, high = 50, 100
lag = numpy.array(corrs[low:high]).argmax()+ low
lag
```

```python
period = lag/saxseg.framerate
period
```

```python
frequency = 1/period
frequency
```

This is approximately a c# 

```python
saxspec = saxseg.make_spectrum()
saxspec.plot()
thinkplot.config(xlabel = 'Frequency', ylabel='Amplitude',xlim=[0,5000],legend=False)
```

```python
saxspec.high_pass(1000, 0.01)
saxspec.plot()
thinkplot.config(xlabel = 'Frequency', ylabel='Amplitude',xlim=[0,5000],legend=False)
```

```python
saxspec.make_wave().make_audio()
```

Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

### analyze1

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy
import math
import pandas
import scipy.fftpack

import dct

PI2 = 2 * math.pi

%matplotlib inline
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
print('analyze1')
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
bests1 = [result.best for result in results]

print('analyze2')
results2 = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze2(ys, freqs, ts)
    results2.append(result)
bests2 = [result2.best for result2 in results2]

print('scipy.fftpack.dct')
results3 = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o scipy.fftpack.dct(ys)
    results3.append(result)
bests3 = [result3.best for result3 in results3]

thinkplot.plot(ns, bests1, label= 'analyze1')
thinkplot.plot(ns, bests2, label = 'analyze2')
thinkplot.plot(ns, bests3, label = 'scipy.fftpack.dct')
thinkplot.config(xscale='log', yscale='log', legend = True, loc = 'upper left', xlabel = 'number of loops', ylabel = 'best run time')
```

Because both anaylze1 and analyze2 increase drastically in the amount of time they take as the number of loops increases, they are both much more complex than scipy.fftpack.dct, which takes about the same amount of time each loop. Analyze1 is the most complex, while analyze 2 follows a similar pattern but takes slightly less time.

## Question 5

One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:

    1. Break a long signal into segments.
    2. Compute the DCT of each segment.
    3. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
    4. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.

Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
catwave = thinkdsp.read_wave('cat_yes.wav')
catwave.make_audio()
```

```python
def dct_iv(ys):
    N = len(ys)
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    args = numpy.outer(ts, freqs)
    M = numpy.cos(PI2 * args)
    amps = numpy.dot(M, ys) / 2
    return amps

def inverse_dct_iv(amps):
    return dct_iv(amps) * 2

def DCTcompression(wave,segment_start,num):
    total = []
    duration = int(wave.duration)
    shortwav = wave.segment(segment_start,1)
    shortspec = shortwav.make_spectrum()
    for a in range(len(shortspec)):
         if shortspec.amps[a] < num:
            shortspec.hs[a]=0
    return shortspec
```
