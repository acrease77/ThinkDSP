## Question 1

1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
N = 10
ts = numpy.arange(N) / N
freqs = numpy.arange(N)
args = numpy.outer(ts, freqs)
M1 = numpy.exp(i * PI2 * args)
```

```python
ts = numpy.arange(N) / N
freqs2 = numpy.arange(N, 2*N)
args2 = numpy.outer(ts, freqs2)
M2 = numpy.exp(i * PI2 * args2)
```

```python
def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol
```

```python
approx_equal(M1, M2)
```

The matricies are exactly the same because they are aliases of eachother. Because the wave is symmetric, it keeps repeating, so values from 0 to N will be identical to values from N to 2N.

## Question 2

2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

```python
sig=thinkdsp.SquareSignal(freq=440)
wave = sig.make_wave(duration=1, framerate=44100)
sig.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
```

```python
spec=wave.make_spectrum()
```

```python
spec2 = wave.make_spectrum()
spec.plot(0.9)
spec2.high_pass(5000)
spec2.plot()
```

```python
amps = spec.amps
amps2 = spec2.amps
ratio = amps2 / amps    
ratio[amps<560] = 0
thinkplot.plot(ratio)

filter1=numpy.ones(len(spec))
for i in range (0,5000):
    filter1[i]=0
        
thinkplot.plot(filter1)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
high_wave = thinkdsp.Wave(filter1, framerate=1)
high_wave.plot()

thinkplot.plot(filter1)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
spec3 = wave.make_spectrum()
spec3.low_pass(5000)

amps = spec.amps
amps3 = spec3.amps
ratio2 = amps3 / amps    
ratio2[amps<560] = 0
thinkplot.plot(ratio2)

filter2=numpy.ones(len(spec))
for i in range (5000,len(spec)):
    filter2[i]=0
        
thinkplot.plot(filter2)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
spec4 = wave.make_spectrum()
spec4.band_stop(5000, 10000)

amps = spec.amps
amps4 = spec4.amps
ratio3 = amps4 / amps    
ratio3[amps<560] = 0
thinkplot.plot(ratio3)

filter3=numpy.ones(len(spec))
for i in range (5000,10000):
    filter3[i]=0
        
thinkplot.plot(filter3)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
def band_pass(spectrum, low, high):
    spectrum.high_pass(low)
    spectrum.low_pass(high)
    return spectrum
```

```python
spec5 = wave.make_spectrum()
band_pass(spec5, 5000, 10000)

amps = spec.amps
amps5 = spec5.amps
ratio4 = amps5 / amps    
ratio4[amps<560] = 0
thinkplot.plot(ratio4)

filter4=numpy.ones(len(spec5))
for i in range (0,5000):
    filter4[i]=0
for i in range (10000,len(spec5)):
    filter4[i]=0
        
thinkplot.plot(filter4)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
high_wave = thinkdsp.Wave(filter1, framerate=1)
high_wave.plot(label='high pass')

low_wave = thinkdsp.Wave(filter2, framerate=1)
low_wave.plot(label='low pass')

stop_wave = thinkdsp.Wave(filter3,framerate=1)
stop_wave.plot(label='band stop')

pass_wave = thinkdsp.Wave(filter4,framerate=1)
pass_wave.plot(label='band pass')

        
thinkplot.plot(filter4)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

When you filter something, it follows a transfer function that describes what you do to the frequencies. You convolve the wndow and the signal in the time domain, so the window is the inverse fourier transform of the filter. Because the frequencies are simply being cut off at certain points, the window, extended to the length of the wave, will be piecewise and look like the filter as a result.

## Question 3

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
sig=thinkdsp.SquareSignal(freq=440)
wave = sig.make_wave(duration=1, framerate=44100)
```

```python
boxcar = numpy.ones(11)
boxcar /= sum(boxcar)

gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
```

```python
ys = numpy.convolve(wave.ys, boxcar, mode='same')
boxcarwave = thinkdsp.Wave(ys, framerate=wave.framerate)

ys2 = numpy.convolve(wave.ys, gaussian, mode='same')
gaussianwave = thinkdsp.Wave(ys2, framerate=wave.framerate)
```

```python
spectrum1 = wave.make_spectrum()
spectrum2 = boxcarwave.make_spectrum()
spectrum3 = gaussianwave.make_spectrum()
amps = spectrum1.amps
amps2 = spectrum2.amps
ratio1 = amps2 / amps    
ratio1[amps<560] = 0

amps3 = spectrum3.amps
ratio2 = amps3 / amps
ratio2[amps<560] = 0
```

```python
angle1 = spectrum1.angles()
angle2 = spectrum2.angles()
angle3 = spectrum3.angles()
thinkplot.plot(spectrum1.fs,angle1)
```

```python
thinkplot.plot(ratio1)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
thinkplot.plot(spectrum2.fs,angle2)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='phase offset', legend=False)
```

```python
thinkplot.plot(ratio2)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim=[0,1.05],
                     legend=False)
```

```python
thinkplot.plot(spectrum3.fs,angle3)
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='phase offset', legend=False)
```

When you look at the plot for the phase offset and frequency in for the boxccar window, you can see that as the "bumps" that the boxcar window creates from convolution go to zero, the phase offset converges around what looks like pi/2 initially, and this value increases up to pi after each consecutive "bump". The phase offset ranges from -pi to pi during a bump, but has a much smaller range inbetween bumps. This is probably due to the convolution of the filter with the spectrum

The Gaussian window also produces some pretty interesting results; it seems that the phase offset of the spectrum converges to a little bit under pi. Once the initial gaussian curve levels to 0, the Gaussian window still creates bumps in its filter from convolution, although they are not as noticeable as the ones from the boxcar window. On the phase and frequency plot, the phase offset generally centers as the curve levels to near zero. The effect that the "bumps" in the Gaussain filter have still appear on the phase offset and frequency plot, although in other places in this plot the range of offsets seems to converge as well. 

## Question 4

Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
siga=thinkdsp.SquareSignal(freq=440)
wave = siga.make_wave(duration=1, framerate=44100)

ys = numpy.convolve(wave.ys, [1,-1], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum = smooth.make_spectrum()
spectrum.plot()
```

```python
ys = numpy.convolve(wave.ys, [-1/2,2,-3/2], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

```python
ys = numpy.convolve(wave.ys, [1,-2,1], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

Both filters from the first approximation [1,-1], and the second [-1/2,2,-3/2] don't really look linear, although they are approaching it. The approximation of the second derivative compared to the first makes some sense, it starts out at 0, with not very much change, because the beginnings of the derivative filters do not change very much, but as the frequency increases they begin to change more rapidly.
