# Question 1

Run and listen to the examples in chap03.ipynb.  In the leakage example, try replacing the Hamming window with one of the other windows provided by NumPy, and see what effect they have on leakage. See http://docs.scipy.org/doc/numpy/reference/routines.window.html

```python
signal = thinkdsp.SinSignal(freq=440)
```

```python
duration = signal.period * 30.5
wave = signal.make_wave(duration)
wave.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
```

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
wave = signal.make_wave(duration)
wave.hamming()
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
def bartlett(self):
    self.ys *=numpy.bartlett(len(self.ys))
    return self

wave = signal.make_wave(duration)
wavebartlett=bartlett(wave)
spectrum = wavebartlett.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
def hanning(self):
    self.ys *=numpy.hanning(len(self.ys))
    return self

wave = signal.make_wave(duration)
wavehanning=hanning(wave)
spectrum = wavehanning.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

The various windows cut off the different frqucny ranges at different amplitudes. While the hamming window seems to cut off the frequencies right as the amplitude begis to drop off, the barlett and hanning windows have some signal leakage, but overall the slope of the frequency decrease on either side looks a bit straighter than the hamming window. The other windows cut off the frequencies at different ranges from the central peak.

# Question 2

Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

Hint: combine the evaluate functions from Chirp and SawtoothSignal.

Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
class ChirpSaw(thinkdsp.Chirp):
    def evaluate(self, ts):
        freqs = numpy.linspace(self.start, self.end, len(ts)-1)
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        
        cycles = numpy.cumsum(dps)/(2*numpy.pi)
        frac, _ = numpy.modf(cycles)        
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
        
```

```python
sawtoothchirp = ChirpSaw(start = 440, end = 880)
wav1 = sawtoothchirp.make_wave(duration=1,framerate = 48000)
wav2 = sawtoothchirp.make_wave(duration=.005, framerate = 48000)
wav2.plot()
wav1.make_audio()
```

```python
spectrum1 = wav2.make_spectrum()
spectrum1.plot()
```

You can hear the aliasing in the overtones and harmonics of the notes. These are visible in the spectrum in the higher ranges.

# Question 3

In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
def glissando(begin,finish):
    gliss1 = thinkdsp.ExpoChirp(start = begin, end = finish)
    wave1 = gliss1.make_wave(duration = 1, framerate = 40000)
    gliss2 = thinkdsp.ExpoChirp(start = finish, end = begin)
    wave2 = gliss2.make_wave(duration = 1, framerate = 40000)
    glissfull = wave1.__or__(wave2)
    return glissfull.make_audio()

glissando(262,349)
```

The length of the trombone is inversely proportional to the pitch. As the length decreases linearly, the pitch increases linearly. Because pitch and frequency have an exponential relationship, as the pitch increases, the frequency increases exponentially. Therefore, over time, the frequency is exponentially increasing if the trombone length decreases, so it is most similar to an exponential chirp.

# Question 6

In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
def BartlettMethod(wave,M,y):
    total = 0
    duration = int(wave.duration)
    for i in range(0,duration):
        shortwav=wave.segment(i*M,M)
        periodogram = shortwav.make_spectrum().hs
        res = periodogram**2/M
        total += res
    bartlett_array=total*M/(duration)
    amp=[]
    for i in bartlett_array:
        amp.append(abs(i))
    thinkplot.plot(amp)
    thinkplot.config(xlabel='frequency (Hz)', ylim=[0, y], ylabel='amplitude', legend=False)
    

BartlettMethod(noisesegment,.1,2000)
```

```python
BartlettMethod(noisesegment,.01,100)
```

Bartlett's method seems to be removing the noise in the system by averaging the frequency values. The plots show Bartlett's method applied with varying values of M. As the M value decrease, the frequency plot becomes less "noisy" are more straight. It gives us a clearer picture of the relationship between frequency and power because it shows the general frequency trend.
