```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%precision 3
%matplotlib inline
```

```python
class ChirpTri(thinkdsp.Chirp):
    def evaluate(self, ts):
        freqs = numpy.linspace(self.start, self.end, len(ts)-1)
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        
        cycles = numpy.cumsum(dps)/(2*numpy.pi)
        frac, _ = numpy.modf(cycles) 
        ys = numpy.abs(frac - 0.5)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
        
```

```python
sawtoothchirp = ChirpTri(start = 1000, end = 1500)
wav1 = sawtoothchirp.make_wave(duration=1,framerate = 48000)
wav2 = sawtoothchirp.make_wave(duration=1, framerate = 48000)
wav2.plot()
```

```python
spectrum1 = wav2.make_spectrum()
spectrum1.plot()
```
